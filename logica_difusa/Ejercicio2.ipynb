```python
import numpy as np

# Funciones de membresía
def pocas_horas(h):
    return max(0, 1 - h / 5)  # Lineal decreciente

def moderadas_horas(h):
    if 3 <= h <= 7:
        return max(0, 1 - abs(h - 5) / 2)
    return 0

def muchas_horas(h):
    return max(0, (h - 5) / 5)

def baja_motivacion(m):
    return max(0, 1 - m / 5)

def media_motivacion(m):
    if 3 <= m <= 7:
        return max(0, 1 - abs(m - 5) / 2)
    return 0

def alta_motivacion(m):
    return max(0, (m - 5) / 5)

# Entrada
H = 5  # Horas de estudio
M = 8  # Nivel de motivación

# Evaluación de reglas
mu_pocas_horas = pocas_horas(H)
mu_moderadas_horas = moderadas_horas(H)
mu_muchas_horas = muchas_horas(H)

mu_baja_motivacion = baja_motivacion(M)
mu_media_motivacion = media_motivacion(M)
mu_alta_motivacion = alta_motivacion(M)

# Aplicación de reglas Takagi-Sugeno
R1 = 30 + 0.5 * H
R2 = 60 + 0.8 * H
R3 = 90 + 0.2 * H

w1 = mu_pocas_horas * mu_baja_motivacion
w2 = mu_moderadas_horas * mu_media_motivacion
w3 = max(mu_muchas_horas, mu_alta_motivacion)

# Cálculo de salida
R = (w1 * R1 + w2 * R2 + w3 * R3) / (w1 + w2 + w3)

# Resultados
print(f"Entrada: H = {H} horas, M = {M} motivación")
print(f"Salida: R = {R:.2f} (calificación estimada)")
```
